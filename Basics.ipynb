#### 3.2.1 Python List slicing
* [start : stop : step ]
  default values start = 0 ; stop = till the last index ; step = 1
* [:stop] will slice from index 0 to the stopping index
* [start : ] will slice from start index to the last index
* Negative values of steps shows the traversal from right to left traversal instead of left to right, so [::-1] reverses the list and prints it


```python
lst=range(1,11)
print(lst)
    
lst1=[1,2,3]
lst1.extend(lst)
print(lst1)

#even numbers
lst_even=lst[1::2]
print(lst_even)

for i in lst_even:
    print(i)

lst_rev=lst[::-1]

for i in lst_rev:
    print(i)

#this prints the even numbers in reverse order
lst_rev_even=lst[9::-2]
for i in lst_rev_even:
    print(i)
```

## 5.0 Control Flow : Loops in Python
* while loop
* while else loop
* different types of **for** loop
    * can also use else for for just like in while
    * **continue** returns the control to the beginning of the loop
    * **break** it brings the control out of the loop
    * **pass** it is used to write empty loops

```python
#while loop in python
def while_loop():
    count=0
    while(count<4):
        print('Hi %d'%(count))
        count=count+1
    print("Printed outside while loop")

if __name__=="__main__":
    while_loop()
```

```python
#while else loop
def while_else():
    count=0
    while(count<4):
        print('Hi %d'%(count))
        count=count+1
    else:
        print("Done printing while and now this is else")

if __name__=="__main__":
    while_else()
```

```python
#for in loop - iterate over tuple, iterate over string and iterate over dictionary
def for_in():
    list1=["Divya","is","smart"]
    for i in list1:
        print(i)
    print("Iterating over string")
    string="Twitter"
    for i in string:
        print(i)
    print("Iterating over Tuple")
    tuple1=('a','b','c')
    for i in tuple1:
        print(i)
    print("Iterating over Dictionary")
    d=dict()
    d['xyz']=100
    d['abc']=200
    for i in d:
        print("%s %d"%(i,d[i]))
    
if __name__=="__main__":
    for_in()
```

### 7.2 The Self
* Class methods must have an extra first parameter in method definition. We do not give value for this parameter when we call the method, Python provides it.
* If we have a method and which takes no arguments, then we still have to have one argument.
* when we call a method of any class, we do not explicitly specify self as the first argument, python takes care of it
* instance variables can be created inside methods also

```python
#an example with self of python
class Human:
    #init method or constructor
    def __init__(self,name):
        self.name=name

    #sample method
    def print_name(self):
        print('My name is',self.name)
        
if __name__=="__main__":
    person=Human("Divya")

    
    person.print_name()
```

```python
#an example to show that instance variables can be created within a method
class Dog:
    animal="dog"
    
    #parameterized constructor
    def __init__(self,breed):
        self.breed=breed   #this is an instance variable
    
    #adds an instance variable
    def set_color(self,color):
        self.color=color
    def get_color(self):
        return self.color

#driver code
brownie=Dog("pug")
brownie.set_color("brown")
print('Brownies color is',brownie.get_color())
print(brownie.color)
```

### 7.3 Constructors and Destructors
* Constructors are used for instantiating an object. Initialise data members of the class when an object of the class is instantiated
    * \_\_init\_\_() is always called whenever an object is created
    * default constructor - no argument except self
    * parameterized constructor - arguments apart from self
* Destructors are called whenever an object gets destroyed. But not much needed in Python as python has better garbage collector when compared to C++
* \_\_del\_\_() is called the destructor method in python and is called when all the references to an object have beem deleted i.e. when an object is garbage collected
    * reference to objects are also deleted when the object goes out of reference or when the program ends
    * destructor is called after the program ends or when the all the references to the objects are deleted and not when object goes out of scope
